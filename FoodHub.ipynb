```python
# import libraries for data manipulation
import numpy as np
import pandas as pd

# import libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('C:/Users/Julia/Downloads/foodhub_order.csv')  # This is where my installed data set is located

# Question 1: Get the shape of the dataset
rows, columns = data.shape

# Display the number of rows and columns
print(f'The dataset contains {rows} rows and {columns} columns.\n')

# Question 2: Get the data types of the columns
data.info()
print('\nSo the data has 1 float, 4 integers, and 4 objects.')

# (This was just to see if it's interesting) Check for missing values and print their count for each column:
print("Number of missing values in each column before treatment (including 0 for numeric):\n")
missing_values_count_before = (data == 0).sum() + data.isnull().sum()  # Consider 0 as missing for numeric columns. 
                                                #I'm assuming no delivery, cook time, cost, etc are ever going to be 0 on purpose. 
                                                #I'm also, possibly incorrectly assuming, the rating system is 1-5
print(missing_values_count_before)

# Print the location of missing values in the dataset (again, just for me because it's interesting)
print("\nLocations of missing values in the dataset:\n")
for column in data.columns:
    missing_indices = data[data[column].isnull()].index
    if not missing_indices.empty:
        print(f"Yes there is missing data: Missing in '{column}' at indices: {list(missing_indices)}")

# Iterate over each column and treat missing values based on its dtype (personal preference)
for column in data.columns:
    if data[column].dtype == 'object':  # If the column is of object type (string)
        data[column].fillna('Not Given', inplace=True)
    elif data[column].dtype == 'int64':  # If the column is an integer
        mode_value = data[column].mode()[0]
        data[column].fillna(mode_value, inplace=True)
    elif data[column].dtype == 'float64':  # If the column is a float
        mean_value = data[column].mean()
        data[column].fillna(mean_value, inplace=True)

# Check for missing values again after treatment
missing_values_after_treatment = data.isnull().sum()
print("\nNumber of missing values in each column after treatment:\n")
print(missing_values_after_treatment)
```
